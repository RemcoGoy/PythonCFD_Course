# Step 1: 1-D Linear Convection

### 1-D Linear Convection Equation

$$ \dfrac{\partial u}{\partial t} + c \dfrac{\partial u}{\partial x} = 0 $$




$$\frac{\partial u}{\partial x} \approx \frac{u(x+\Delta x) - u(x)}{\Delta x}$$
$$\frac{u_i^{n+1} - u_i^n}{\Delta t} + c\frac{u_i^n - u_{i-1}^n}{\Delta x} = 0$$
$$u_i^{n+1} = u_i^n - c\frac{\Delta t}{\Delta x}(u_i^n - u_{i-1}^n)$$

# Python Implementation
## Imports

```python
import numpy as np
from matplotlib import pyplot
import time, sys

%matplotlib inline
```

```python
nx = 61 # Number of grid points
dx = 2 / (nx - 1) # Grid spacing (delta x)
nt = 20 # Number of time steps
dt = 0.025 # Time step (delta t)
c = -1 # Wave speed
u = np.ones(nx) # Initial condition
u[int(0.5 / dx):int(1 / dx + 1)] = 2 # Set hat function in the center of the domain
```

```python
pyplot.plot(np.linspace(0, 2, nx), u);
```

```python
un = np.ones(nx) # Initialize a temporary array
for n in range(nt): # Loop over the time steps
    un = u.copy() # Copy the existing values of u into un
    for i in range(1, nx): # Loop over the grid points
        u[i] = un[i] - c * dt / dx * (un[i] - un[i - 1])
```

```python
pyplot.plot(np.linspace(0, 2, nx), u);
```
